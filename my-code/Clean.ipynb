```python
# edit Sex column
# new list with valid entries

F_M_sex = ['F', 'M']
```

```python
# create a new column with edited Sex entries, adding 'unknown' for all entries that were not correctly defined
# rename the column and remove space _

sharks['Sex'] = sharks['Sex '].apply(lambda x: 'unknown' if x not in F_M_sex else x)
```

```python
# Type: drop everything that's invalid

invalid_type = ['Invalid', 'Questionable', 'Boatomg']
```

```python
sharks = sharks[sharks.Type.str.contains('Invalid') == False]
sharks = sharks[sharks.Type.str.contains('Questionable') == False]
sharks = sharks[sharks.Type.str.contains('Boatomg') == False]

# not sure how I can pass the whole list in contains (?) 
```
